# Days of Week Stats

```python
import datetime
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

filename = "@GC_60min.txt"

startDate = "20070810"
endDate = "20180810"
startDateParsed = datetime.date(int(startDate[0:4]), int(startDate[4:6]), int(startDate[6:]))
endDateParsed = datetime.date(int(endDate[0:4]), int(endDate[4:6]), int(endDate[6:]))

instrument = pd.read_csv(filename, 
                   usecols=['Date','Time','Open','High','Low','Close'], parse_dates=[['Date', 'Time']])
print("Il file contiene", len(instrument), "record")
instrument.index = instrument['Date_Time']
instrument.drop('Date_Time',axis=1,inplace=True)
instrument['AvgPrice'] = round(instrument.iloc[:,0:4].mean(axis=1),2)
instrument['OpenClose'] = round((instrument['Close'] - instrument['Open']),2)
instrument['OpenClosePerc'] = round((instrument['Close'] - instrument['Open']) / instrument['Open'] * 100,2)
instrument['DeltaClosePerc'] = round((instrument['Close'] - instrument['Close'].shift(1)) / instrument['Close'].shift(1) * 100,2)
instrument['DayOfMonth'] = instrument.index.day
instrument['DayOfWeek'] = instrument.index.dayofweek
instrument['Week'] = instrument.index.week
instrument['Month'] = instrument.index.month

instrument = instrument.loc[startDateParsed:endDateParsed]
instrument.head(24)
```

Aggiungiamo la colonna "DailyOpenClosePerc" in cui memorizziamo la somma dei contributi giornalieri di "OpenClosePerc"

```python
daily = instrument.resample('D').mean()
daily['DailyOpenClosePerc'] = instrument.resample('D')['OpenClosePerc'].sum()
daily['DailyDeltaClosePerc'] = instrument.resample('D')['DeltaClosePerc'].sum()
daily.head(10)
```

```python
daily['DailyOpenClosePerc'].describe(percentiles = [0.10,0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90])
```

```python
daily['DailyDeltaClosePerc'].describe(percentiles = [0.10,0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90])
```

```python
biasWeeklyIntraday = []
biasWeeklyOverNight = []
daysofweek = []

for day in range(0,7):
    daysofweek.append(day)

print(daysofweek)

for day in daysofweek:
    #print(dayofweek)
    biasWeeklyIntraday.append(daily[(daily['DayOfWeek']==day)]['DailyOpenClosePerc'].mean())
    biasWeeklyOverNight.append(daily[(daily['DayOfWeek']==day)]['DailyDeltaClosePerc'].mean())

print("")
print("Intraday:")
print(len(biasWeeklyIntraday))
print(biasWeeklyIntraday)
print("")
print("OverNight:")
print(len(biasWeeklyOverNight))
print(biasWeeklyOverNight)
```

```python
# ************** MODULO GRAFICO **********************************************************************************
n_groups = 7

fig, ax = plt.subplots(figsize=(14, 8), dpi=300)

index = np.arange(n_groups)
bar_width = 0.35
opacity = 1

rects1 = ax.bar(index, 
                biasWeeklyIntraday, 
                bar_width, 
                alpha=opacity, 
                color='red', 
                label='Intraday Day of Week Seasonality')

rects1 = ax.bar(index + bar_width, 
                biasWeeklyOverNight, 
                bar_width, 
                alpha=opacity, 
                color='blue', 
                label='Overnight Day of Week Seasonality')

ax.set_xlabel('Day of Week')
ax.set_ylabel('Average Percent Gain')
ax.set_title('Day of Week Seasonality ' + ' - by Gandalf Project R&D')
ax.set_xticks(index)
ax.set_xticklabels((["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]))
ax.legend()
ax.grid(True)
plt.show()
# ************** MODULO GRAFICO **********************************************************************************
```

# Percentuale occorrenze positive e negative

Creaimo una funzione che restituisca la percentuale di valori al di sopra e al di sotto di una soglia:

```python
def percPosNeg(array, soglia):
    """
    Funzione che restituisce le percentuali di valori 
    maggiori e minori di una soglia
    """
    if len(array) == 0:
        return 0,0
    contapos = 0
    contaneg = 0
    for el in array:
        if el > soglia:
            contapos+=1
        if el <= soglia:
            contaneg+=1
    return round(contapos/len(array)*100,2), round(contaneg/len(array)*100,2)

DayOfWeek_Intraday_Statistics = []
DayOfWeek_OverNight_Statistics = []

daysofweek_names = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

for day in daysofweek:

    intraday = daily[(daily['DayOfWeek']==day)]['DailyOpenClosePerc'].dropna()
    DayOfWeek_Intraday_Statistics.append(percPosNeg(intraday,0))
    
    overnight = daily[(daily['DayOfWeek']==day)]['DailyDeltaClosePerc'].dropna()
    DayOfWeek_OverNight_Statistics.append(percPosNeg(overnight,0))

print("Statistics Intraday:")
for el in range(len(DayOfWeek_Intraday_Statistics)):
    if biasWeeklyIntraday[el] > 0:
        print(daysofweek_names[el], round(biasWeeklyIntraday[el],4), DayOfWeek_Intraday_Statistics[el][0], "%")
    if biasWeeklyIntraday[el] <= 0:
        print(daysofweek_names[el], round(biasWeeklyIntraday[el],4), DayOfWeek_Intraday_Statistics[el][1], "%")
        
print("")
print("Statistics Overnight:")
for el in range(len(DayOfWeek_OverNight_Statistics)):
    if biasWeeklyOverNight[el] > 0:
        print(daysofweek_names[el], round(biasWeeklyOverNight[el],4), DayOfWeek_OverNight_Statistics[el][0], "%")
    if biasWeeklyOverNight[el] <= 0:
        print(daysofweek_names[el], round(biasWeeklyOverNight[el],4), DayOfWeek_OverNight_Statistics[el][1], "%")
```

```python
import datetime
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

def percPosNeg(array, soglia):
    """
    Funzione che restituisce le percentuali di valori 
    maggiori e minori di una soglia
    """
    if len(array) == 0:
        return 0,0
    contapos = 0
    contaneg = 0
    for el in array:
        if el > soglia:
            contapos+=1
        if el <= soglia:
            contaneg+=1
    return round(contapos/len(array)*100,2), round(contaneg/len(array)*100,2)

filename = "@GC_60min.txt"

startDate = "20070810"
endDate = "20180810"
startDateParsed = datetime.date(int(startDate[0:4]), int(startDate[4:6]), int(startDate[6:]))
endDateParsed = datetime.date(int(endDate[0:4]), int(endDate[4:6]), int(endDate[6:]))

instrument = pd.read_csv(filename, 
                   usecols=['Date','Time','Open','High','Low','Close'], parse_dates=[['Date', 'Time']])
print("Il file contiene", len(instrument), "record")
instrument.index = instrument['Date_Time']
instrument.drop('Date_Time',axis=1,inplace=True)
instrument['AvgPrice'] = round(instrument.iloc[:,0:4].mean(axis=1),2)
instrument['OpenClose'] = round((instrument['Close'] - instrument['Open']),2)
instrument['OpenClosePerc'] = round((instrument['Close'] - instrument['Open']) / instrument['Open'] * 100,2)
instrument['DeltaClosePerc'] = round((instrument['Close'] - instrument['Close'].shift(1)) / instrument['Close'].shift(1) * 100,2)
instrument['DayOfMonth'] = instrument.index.day
instrument['DayOfWeek'] = instrument.index.dayofweek
instrument['Week'] = instrument.index.week
instrument['Month'] = instrument.index.month

instrument = instrument.loc[startDateParsed:endDateParsed]
instrument.head(24)

daily = instrument.resample('D').mean()
daily['DailyOpenClosePerc'] = instrument.resample('D')['OpenClosePerc'].sum()
daily['DailyDeltaClosePerc'] = instrument.resample('D')['DeltaClosePerc'].sum()
daily.head(10)

print("Statistiche Intraday:")
print(daily['DailyOpenClosePerc'].describe(percentiles = [0.10,0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90]))
print("")
print("Statistiche OverNight:")
print(daily['DailyDeltaClosePerc'].describe(percentiles = [0.10,0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90]))

daysofweek = []
months = []

for day in range(0,7):
    daysofweek.append(day)
    
for month in range(1,13):
    months.append(month)
    
months_names = ["January",
                "February",
                "March",
                "April",
                "May",
                "June",
                "July",
                "August",
                "September",
                "October",
                "November",
                "December"]

print(daysofweek)
print(months)

daysofweek_names = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

for month in months:
    biasWeeklyIntraday = []
    biasWeeklyOverNight = []
    DayOfWeek_Intraday_Statistics = []
    DayOfWeek_OverNight_Statistics = []
    for day in daysofweek:
        biasWeeklyIntraday.append(daily[(daily['DayOfWeek']==day) & (daily['Month']==month)]['DailyOpenClosePerc'].mean())
        biasWeeklyOverNight.append(daily[(daily['DayOfWeek']==day) & (daily['Month']==month)]['DailyDeltaClosePerc'].mean())
        
        intraday = daily[(daily['DayOfWeek']==day) & (daily['Month']==month)]['DailyOpenClosePerc'].dropna()
        DayOfWeek_Intraday_Statistics.append(percPosNeg(intraday, 0))
        overnight = daily[(daily['DayOfWeek']==day) & (daily['Month']==month)]['DailyDeltaClosePerc'].dropna()
        DayOfWeek_OverNight_Statistics.append(percPosNeg(overnight, 0))

    print("")
    print("Statistics Intraday:")
    for el in range(len(DayOfWeek_Intraday_Statistics)):
        if biasWeeklyIntraday[el] > 0:
            print(daysofweek_names[el], round(biasWeeklyIntraday[el], 4), DayOfWeek_Intraday_Statistics[el][0], "%")
        if biasWeeklyIntraday[el] <= 0:
            print(daysofweek_names[el], round(biasWeeklyIntraday[el], 4), DayOfWeek_Intraday_Statistics[el][1], "%")

    print("")
    print("Statistics Overnight:")
    for el in range(len(DayOfWeek_OverNight_Statistics)):
        if biasWeeklyOverNight[el] > 0:
            print(daysofweek_names[el], round(biasWeeklyOverNight[el], 4), DayOfWeek_OverNight_Statistics[el][0], "%")
        if biasWeeklyOverNight[el] <= 0:
            print(daysofweek_names[el], round(biasWeeklyOverNight[el], 4), DayOfWeek_OverNight_Statistics[el][1], "%")

    # ************** MODULO GRAFICO **********************************************************************************
    n_groups = 7

    fig, ax = plt.subplots(figsize=(14, 8), dpi=300)

    index = np.arange(n_groups)
    bar_width = 0.35
    opacity = 1

    rects1 = ax.bar(index, 
                    biasWeeklyIntraday, 
                    bar_width, 
                    alpha=opacity, 
                    color='red', 
                    label='Intraday Day of Week Seasonality')

    rects1 = ax.bar(index + bar_width, 
                    biasWeeklyOverNight, 
                    bar_width, 
                    alpha=opacity, 
                    color='blue', 
                    label='Overnight Day of Week Seasonality')

    ax.set_xlabel('Day of Week')
    ax.set_ylabel('Average Percent Gain')
    ax.set_title('Day of Week Seasonality ' + 'on ' + months_names[month-1] + ' - by Gandalf Project R&D')
    ax.set_xticks(index)
    ax.set_xticklabels((["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]))
    ax.legend()
    ax.grid(True)
    plt.show()
    # ************** MODULO GRAFICO **********************************************************************************
```
