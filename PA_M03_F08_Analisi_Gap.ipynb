# Prendiamo una serie azionaria daily:

Carichiamo la serie dei prezzi e calcoliamo i GAP e i LAP nominali e percentuali:

```python
import datetime
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

print(__version__) # necessaria versione >= 1.9.0

import cufflinks as cf

# Per utilizzo con Notebooks
#init_notebook_mode(connected=True)

# Per utilizzo offline
cf.go_offline()

# INPUTS ****************************
filename = "AAPL_Daily.txt"

startDate = "20070810"
endDate =   "20180810"

# INPUTS ****************************
    
startDateParsed = datetime.date(int(startDate[0:4]), int(startDate[4:6]), int(startDate[6:]))
endDateParsed = datetime.date(int(endDate[0:4]), int(endDate[4:6]), int(endDate[6:]))

instrument = pd.read_csv(filename, 
                   usecols=['Date','Time','Open','High','Low','Close'], parse_dates=[['Date', 'Time']])
print("Il file contiene", len(instrument), "record")
instrument.index = instrument['Date_Time']
instrument.drop('Date_Time',axis=1,inplace=True)
instrument['AvgPrice'] = round(instrument.iloc[:,:4].mean(axis=1),2)
instrument['CloseOpen'] = round((instrument['Close'] - instrument['Open']),2)
instrument['CloseOpenPerc'] = round((instrument['Close'] - instrument['Open']) / instrument['Open'] * 100,2)
#instrument['DeltaClosePerc'] = round((instrument['Close'] - instrument['Close'].shift(1)) / instrument['Close'].shift(1) * 100,2)

instrument['Delta'] = instrument['Open'] - instrument['Close'].shift(1)

instrument['GapUp'] = np.where((instrument['Open'] - instrument['High'].shift(1)) > 0, 
                               (instrument['Open'] - instrument['High'].shift(1)), 
                               np.nan)

instrument['GapDown'] = np.where((instrument['Open'] - instrument['Low'].shift(1)) < 0,
                                 (instrument['Open'] - instrument['Low'].shift(1)), np.nan)

instrument['LapUp'] = np.where((instrument['Open'] > instrument['Close'].shift(1)) & \
                               (instrument['Open'] <= instrument['High'].shift(1)), 
                               (instrument['Open'] - instrument['Close'].shift(1)), np.nan)

instrument['LapDown'] = np.where((instrument['Open'] < instrument['Close'].shift(1)) & \
                                 (instrument['Open'] >= instrument['Low'].shift(1)), 
                                 (instrument['Open'] - instrument['Close'].shift(1)), np.nan)

instrument = instrument.loc[startDateParsed:endDateParsed]

instrument.head(20)
```
